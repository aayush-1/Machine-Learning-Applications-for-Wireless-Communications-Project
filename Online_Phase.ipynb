```python
from keras.models import Sequential
from keras.utils import np_utils
from keras.layers.normalization import BatchNormalization
import cmath
from sklearn import metrics 
from keras.layers.core import Activation
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from keras.layers import Dense, Conv2D, MaxPooling2D, Dropout, Flatten
from sklearn.model_selection import train_test_split
import tensorflow as tf
from keras.layers import Dense, Input
from keras.models import Model
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=pd.read_csv('TestData.csv')
print(data)
xtest=data.values[:,1:]
print(xtest.shape)
```

```python
#Compute the variance of CSI values
variance=np.var(xtest)
print(variance)
```

```python
from keras.models import load_model
v=xtest


p=[]
for i in range(2):
    v_bar=np.zeros((v.shape))
    #print(v)
    #print(v_bar.shape)
    for j in range(4):
        model=load_model("model_weights_link"+str(j+1)+"_locn"+str(i+1)+".h5")
        a=v[:,j%4]
        a=a.reshape(1,-1)
        model.summary()
        v_bar[:,j%4]=model.predict(a)
    v_bar=1/(1+np.exp(-v_bar))
    p.append(np.exp((-1/variance)*np.sqrt((np.linalg.norm(v-v_bar))**2)))
#calculation of prosterior probability
print(p[0]/(p[0]+p[1]))
print(p[1]/(p[0]+p[1]))

#the position depends on the higher probability (locn-1 or locn-2)



        
            
        

    
    
```
