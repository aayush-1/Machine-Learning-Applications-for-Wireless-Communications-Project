```python
from keras.models import Sequential
from keras.utils import np_utils
from keras.layers.normalization import BatchNormalization
import cmath
from sklearn import metrics 
from keras.layers.core import Activation
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from keras.layers import Dense, Conv2D, MaxPooling2D, Dropout, Flatten
from sklearn.model_selection import train_test_split
import tensorflow as tf
from keras.layers import Dense, Input
from keras.models import Model


```

```python
data=pd.read_csv('data.csv')
print(data)
```

```python
# Link-1 Location-1 Autoencoder

inp_link1_locn1=data.values[:30,1]
inp_link1_locn1=inp_link1_locn1.reshape(1,-1)
#print(inp.shape)
inp_link1_locn1 = tf.convert_to_tensor(inp_link1_locn1, np.float32)


inputs = Input(shape=(30,))

layer_1 = Dense(15, activation='relu')
layer_1_output = layer_1(inputs)

layer_2 = Dense(15, activation='relu')
layer_2_output = layer_2(layer_1_output)

transpose_layer = Dense(30, activation='relu', use_bias=False)
transpose_layer_output = transpose_layer(layer_2_output)

transpose_layer.set_weights([layer_1.get_weights()[0].T])

model_link1_locn1 = Model(inputs=inputs, outputs=transpose_layer_output)
model_link1_locn1.compile('adam', loss='mean_squared_error')
model_link1_locn1.summary()

history_link1_locn1 = model_link1_locn1.fit(inp_link1_locn1, inp_link1_locn1,steps_per_epoch=1, epochs=100
                                            , verbose=2)
model_link1_locn1.save_weights('model_weights_link1_locn1.h5')
```
